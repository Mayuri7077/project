## splitting dataset into train set and test set

```python
X = df.values
X = np.delete(X,9,axis=1)
Y = df['y'].values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(Y)
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 1)
```

## fitting model

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
# Decision Tree
model1 = DecisionTreeClassifier(max_depth=5)

# fit model
model1.fit(X_train, Y_train)

# prediction
y_pred1 = model1.predict(X_test)

# predict probabilities
pred_prob1 = model1.predict_proba(X_test)

# classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test, y_pred1))
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier 

# Random Forest Classifier
model2 = RandomForestClassifier(n_estimators=100)

# fit model
model2.fit(X_train, Y_train)

# prediction
y_pred2 = model2.predict(X_test)

# predict probabilities
pred_prob2 = model2.predict_proba(X_test)

# classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test, y_pred2))
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

# Random Forest Classifier
model3 = LogisticRegression()

# fit model
model3.fit(X_train, Y_train)

# prediction
y_pred3 = model3.predict(X_test)

# predict probabilities
pred_prob3 = model3.predict_proba(X_test)

# classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test, y_pred3))
```

K Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

# K Neighbors Classifier
model4 = KNeighborsClassifier(n_neighbors=4)

# fit model
model4.fit(X_train, Y_train)

# prediction
y_pred4 = model4.predict(X_test)

# predict probabilities
pred_prob4 = model4.predict_proba(X_test)

# classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test, y_pred4))
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

# Naive Bayes
model5 = GaussianNB()

# fit model
model5.fit(X_train, Y_train)

# prediction
y_pred5 = model5.predict(X_test)

# predict probabilities
pred_prob5 = model5.predict_proba(X_test)

# classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test, y_pred5))
```

##AUC-ROC curve

```python
from sklearn.metrics import roc_curve

# roc curve for models
fpr1, tpr1, thresh1 = roc_curve(Y_test, pred_prob1[:,1], pos_label=1)
fpr2, tpr2, thresh2 = roc_curve(Y_test, pred_prob2[:,1], pos_label=1)
fpr3, tpr3, thresh3 = roc_curve(Y_test, pred_prob3[:,1], pos_label=1)
fpr4, tpr4, thresh4 = roc_curve(Y_test, pred_prob4[:,1], pos_label=1)
fpr5, tpr5, thresh5 = roc_curve(Y_test, pred_prob5[:,1], pos_label=1)
#fpr6, tpr6, thresh6 = roc_curve(Y_test, pred_prob6[:,1], pos_label=1)

# roc curve for tpr = fpr 
random_probs = [0 for i in range(len(Y_test))]
p_fpr, p_tpr, _ = roc_curve(Y_test, random_probs, pos_label=1)
```

```python
from sklearn.metrics import roc_auc_score

# auc scores
auc_score1 = roc_auc_score(Y_test, pred_prob1[:,1])
auc_score2 = roc_auc_score(Y_test, pred_prob2[:,1])
auc_score3 = roc_auc_score(Y_test, pred_prob3[:,1])
auc_score4 = roc_auc_score(Y_test, pred_prob4[:,1])
auc_score5 = roc_auc_score(Y_test, pred_prob5[:,1])
#auc_score6 = roc_auc_score(Y_test, pred_prob6[:,1])


print("Decision Tree ROC-AUC score: ",auc_score1)
print("Random Forest Classifier ROC-AUC score: ",auc_score2)
print("Logistic Regression ROC-AUC score: ",auc_score3)
print("K Nearest Neighbor ROC-AUC score: ",auc_score4)
print("Naive Bayes ROC-AUC score: ",auc_score5)
#print("Support Vector Machine ROC-AUC score: ",auc_score6)
```

```python
# matplotlib
import matplotlib.pyplot as plt
plt.style.use('seaborn')

# plot roc curves
plt.plot(fpr1, tpr1, linestyle='--',color='orange', label='Decision Tree')
plt.plot(fpr2, tpr2, linestyle='--',color='green', label='Random Forest')
plt.plot(fpr3, tpr3, linestyle='--',color='red', label='Logistic Regression')
plt.plot(fpr4, tpr4, linestyle='--',color='pink', label='KNN')
plt.plot(fpr5, tpr5, linestyle='--',color='yellow', label='Naive Bayes')
#plt.plot(fpr6, tpr6, linestyle='--',color='purple', label='Support Vector Machine')

plt.plot(p_fpr, p_tpr, linestyle='--', color='blue')

# title
plt.title('ROC curve')

# x label
plt.xlabel('False Positive Rate')

# y label
plt.ylabel('True Positive rate')

plt.legend(loc='best')
plt.savefig('ROC',dpi=300)
plt.show();
```
